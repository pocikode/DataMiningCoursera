```python
import pandas as pd
import matplotlib.pyplot as plt
from aquarel import load_theme
```

```python
theme = load_theme("umbra_dark")
theme.apply()

filepath = 'ExcelFormattedGISTEMPData2CSV.csv'
df = pd.read_csv(filepath)
df.head()
```

```python
plt.figure(figsize=(12, 7))
plt.plot(df['Year'], df['Glob'], label='Global Temperature (Glob)')
plt.plot(df['Year'], df['NHem'], label='Northern Hemisphere Temperature (NHem)')
plt.plot(df['Year'], df['SHem'], label='Southern Hemisphere Temperature (SHem)')

plt.title('Global, Northern and Southern Hemisphere Temperature Changes (1880-2014)', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Temperature Change (°C)', fontsize=14)
plt.legend(fontsize=12)

plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
theme.apply_transforms()

plt.show()
```
